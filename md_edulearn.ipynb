#Tabela Aluno
Organização e tratamento dos dados para criação de arquivo csv para carregamento dos dados da tabela Aluno no banco de dados MySQL

```python
#Importando os dados
df_aluno = pd.read_csv('alunodeficiente_2019.csv', sep=',', usecols=['CO_IES','TP_COR_RACA','NU_ANO_NASCIMENTO','ID_ALUNO','CO_PAIS_ORIGEM','CO_UF_NASCIMENTO','CO_MUNICIPIO_NASCIMENTO','TP_SEXO'])
df_aluno.head()
```

```python
#Renomeando as colunas
df_aluno = df_aluno.rename(columns={
    'CO_IES':'CODIGO_IES',
    'ID_ALUNO':'CODIGO_INEP',
    'TP_COR_RACA':'RACA',
    'TP_SEXO':'GENERO',
    'CO_PAIS_ORIGEM':'PAIS_DE_ORIGEM',
    'CO_UF_NASCIMENTO':'COD_UF',
    'CO_MUNICIPIO_NASCIMENTO':'COD_MUNICIPIO',
    })
df_aluno.head()
```

```python
#Substituindo o código pela raça
raca = {
    '0':'Aluno não quis declarar cor/raça',   
    '1':'Branca',
    '2':'Preta',
    '3':'Parda',
    '4':'Amarela',
    '5':'Indígena',
    '9':'Não dispõe da informação',
}
df_aluno = df_aluno.astype({'RACA':str})
df_aluno.RACA = df_aluno.RACA.map(raca)
df_aluno.head()
```

```python
#Substituindo código por gênero
genero = {
    '1':'Feminino',
    '2':'Masculino',
}
df_aluno = df_aluno.astype({'GENERO':str})
df_aluno.GENERO = df_aluno.GENERO.map(genero)
df_aluno.head()
```

```python
ano_atual = dt.date.today().year
df_aluno['ANO_ATUAL'] = ano_atual
df_aluno.head()
```

```python
#Inserindo a coluna idade do aluno
df_aluno = df_aluno.astype({'ANO_ATUAL':int})
df_aluno = df_aluno.astype({'NU_ANO_NASCIMENTO':int})
df_aluno['IDADE'] = df_aluno['ANO_ATUAL'] - df_aluno['NU_ANO_NASCIMENTO']
df_aluno.head()
```

```python
#Verificando se há CODIGO_INEP duplicaado
df_aluno['CODIGO_INEP'].duplicated()
```

Observação: Como há códigos do inep duplicados para alunos com dados diferentes, resolvemos não considerar essa coluna como identificadora e colocar uma id AI no banco de dados como chave primaria do aluno.

```python
#Excluindo a coluna código inep
df_aluno = df_aluno.drop(columns=['CODIGO_INEP'])
```

```python
#Para referenciar o código da IES com o id da IES do banco de dados
id_ies = pd.read_csv('tabela_com_id_ies.csv')
id_ies.head()
```

```python
#Dando um join com o código da IES
df_aluno = df_aluno.join(id_ies[['id_ies', 'cod']].set_index('cod'), on='CODIGO_IES')
df_aluno.head()
```

```python
df_paises = pd.read_csv('codigo_pais.csv', sep=',')
df_paises.head()
```

```python
df_aluno = df_aluno.join(df_paises[['CO_PAIS', 'NO_PAIS']].set_index('CO_PAIS'), on='PAIS_DE_ORIGEM')
df_aluno.head()
```

```python
df_municipios = pd.read_csv('municipio.csv', sep=',', encoding='latin-1')
df_municipios.head()
```

```python
df_aluno = df_aluno.join(df_municipios[['COD_MUNICIPIO', 'MUNICIPIO']].set_index('COD_MUNICIPIO'), on='COD_MUNICIPIO')
df_aluno.head()
```

```python
df_aluno = df_aluno.drop(['PAIS_DE_ORIGEM','COD_MUNICIPIO'], axis=1)
df_aluno.head()
```

```python
#Organizando as colunas do dataframe
df_aluno = df_aluno.rename(columns={
    'NO_PAIS':'PAIS_ORIGEM',
    })
df_aluno.head()
```

```python
#Inserindo a id do aluno a partir do index do dataframe
df_aluno.index = range(1,len(df_aluno.index)+1)
df_aluno['ID_ALUNO'] = df_aluno.index
df_aluno.head()
```

```python
#Organizando as colunas do dataframe
df_aluno = df_aluno[['ID_ALUNO','CODIGO_IES','RACA','PAIS_ORIGEM','GENERO','IDADE']]
df_aluno.head()
```

```python
#Exportando o dataframe para csv
df_aluno.to_csv('tabela_aluno.csv', index=False, sep=';')
```
